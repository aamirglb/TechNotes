```python
width = 5
height = 4
img = np.zeros([height, width, 3], dtype=np.uint8)
img[:, :] = [0xFF, 0x7F, 0]
plt.imshow(img)
```

```python
img2 = np.zeros([100, 200, 3], dtype=np.uint8)
img2[: , :100] = [0xFF, 0x7F, 0]
img2[: , 100:] = [0x0, 0x0, 0xFF]
plt.imshow(img2)
```

```python
img3 = np.zeros([200, 200, 3], dtype=np.uint8)
img3[:100 , :] = [0xFF, 0x7F, 0x0]
img3[100: , :] = [0x0, 0x0, 0xFF]
plt.imshow(img3)
```

```python
img4 = np.zeros([100, 100, 3], dtype=np.uint8)
img4[:50 , :50] = [0xFF, 0x0, 0x0]
img4[:50 , 50:] = [0x0, 0xFF, 0x0]
img4[50: , :50] = [0x0, 0x0, 0xFF]
img4[50: , 50:] = [0xFF, 0xFF, 0x0]
plt.imshow(img4)
```

```python
img5 = np.zeros([100, 200, 4], dtype=np.uint8)
img5[:, :100] = [0xFF, 0x7F, 0x0, 0xFF]
img5[:, 100:] = [0x0, 0x0, 0xFF, 0xFF]
for x in range(200):
    for y in range(100):
        img5[y, x, 3] = x
plt.imshow(img5)
```

```python
img6 = np.zeros([100, 200, 4], dtype=np.uint8)
img6[:, :100] = [0xFF, 0x7F, 0x0, 0xFF]
img6[:, 100:] = [0x0, 0x0, 0xFF, 0xFF]
for x in range(200):
    for y in range(100):
        img6[y, x, 3] = y+50
plt.imshow(img6)
```

```python
img = np.zeros([100, 200], dtype=np.uint8)
for x in range(200):
    for y in range(100):
            img[y, x] = x
plt.imshow(img, cmap='gray')
```

```python
img7 = np.zeros([100, 200], dtype=np.uint8)
for x in range(200):
    for y in range(100):
        if (x % 16) // 8 == (y % 16) // 8:
            img7[y, x] = 0
        else:
            img7[y, x] = 0xFF
plt.imshow(img7, cmap='gray')
```

## Splittin and Merging Color Channels

```python
img_nz = cv.imread('/home/aamir/Pictures/nz_lake.jpg', cv.IMREAD_COLOR)
b,g,r = cv.split(img_nz)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title('Red channel')
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title('Green channel')
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title('Blue channel')

img_all = cv.merge((b,g,r))
plt.subplot(144);plt.imshow(img_all[:,:,::-1]);plt.title('Merged')
```

```python
img_bgr = cv.imread('/home/aamir/Pictures/nz_lake.jpg', 1)
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(img_bgr);plt.title('BGR')
plt.subplot(142);plt.imshow(img_rgb);plt.title('RGB')
# plt.imshow(img_bgr)
# plt.imshow(img_rgb)
```
