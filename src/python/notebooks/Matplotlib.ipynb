# Matplotlib

The most popular general-purpose visualization library
is **Matplotlib**, (www.matplotlib.org) which mainly focuses on generating static publication-quality 2D and 3D graphs. Many other libraries focus on niche areas of visualization. A few prominent
examples are 

1. Bokeh (http://bokeh.pydata.org) and Plotly (http://plot.ly), which
both primarily focus on interactivity and web connectivity, 
2. Seaborn (http://stanford.edu/~mwaskom/software/seaborn) which is a high-level plotting library which targets statistical data analysis and which is based on the Matplotlib library, and the 
3. Mayavi library (http://docs.enthought.com/mayavi/mayavi) for high-quality 3D visualization, which uses the venerable VTK software (http://www.vtk.org) for heavy-duty scientific visualization
4. VisPy (http://vispy.org), which is an OpenGL-based 2D and 3D visualization library with great interactivity and connectivity with browser-based environments, such as the Jupyter Notebook.

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

import numpy as np
import sympy
mpl.__version__, np.__version__, sympy.__version__
```

Matplotlib actually provides multiple entry points into the
library, with different application programming interfaces (APIs). Specifically, it provides a _stateful_ API and an _object-oriented_ API, both provided by the module `matplotlib.pyplot`.

A graph in **Matplotlib** is structured in terms of a `Figure` instance and one or more `Axes` instances within the figure. The `Figure` instance provides a canvas area for drawing, and the `Axes` instances provide coordinate systems that are assigned to fixed regions of the total figure canvas.

The Axes instance provides a coordinate system that can be used to plot data in a variety of plot styles, including line graphs, scatter plots, bar plots, and many other styles.

$y(x) = x^(3)+5x^2+10$

```python
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10

fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label="y(x)")
ax.plot(x, y2, color='red', label="y'(x)")
ax.plot(x, y3, color='green', label="y\"(x)")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

plt.subplots function generate Figure and Axes instances.

```python
x = np.linspace(0, np.pi*4, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, ax = plt.subplots()
ax.plot(x, y1, color="blue", label="sin(x)")
ax.plot(x, y2, color="red", label="cos(x)")
# ax.plot(x, y3, color="green", label="tan(x)")
ax.legend()
```

```python
x = np.linspace(-1, 1, 100)
y1 = x**2
y2 = x**3
# y3 = x**0.5
y4 = x**4

fig, ax = plt.subplots()
ax.plot(x, y1, color="blue", label="x**2")
ax.plot(x, y2, color="green", label="x**3")
# ax.plot(x, y3, color="red", label="x**0.5")
ax.plot(x, y4, color="gray", label="x**4")
ax.legend()
```

Cubic Function

$f(x) = (x^3 + 3x^2 - 6x - 8)/4$

```python
x = np.linspace(-45, 45, 200)
y = (x**3 + 3*x**2 - 6*x - 8)/4

fig, ax = plt.subplots()
ax.plot(x, y, color="red", label="f(x)")
ax.legend()
```
