# Script to Extract Mavlink IDs and Names
This script read and parse the message definitions XML and generate C++ map entries to be used in a C++ program.

```python
import os
import xml.etree.ElementTree as ET

# Directory where all message definitions XML are stored
input_dir = '/home/strio/worksapce/mavlink/include/mavlink/v2.0/message_definitions'
xml_files = []
(dirpath, dirnames, xml_files) = next(os.walk(input_dir))

cnt = 1
for i, f in enumerate(xml_files):
    xml_file = dirpath + "/" + f
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    for child in root:
        if child.tag == 'messages':
            msgs = child.findall('message')            
            for m in msgs:
                e = '\n' if (cnt % 3) == 0 else ''
                s = '{{ {}, "{}" }},'.format(m.attrib['id'], m.attrib['name'])
                # Generate C++ map entries
                print('{0: <40}'.format(s), end=e)
                cnt+=1
```
